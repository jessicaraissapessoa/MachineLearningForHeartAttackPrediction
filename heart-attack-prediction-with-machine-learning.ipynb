# Importações e leitura de CSV

```python
# Bibliotecas para manipulação e computação de dados
import pandas as pd  # Manipulação e análise de dados tabulares, permite leitura de arquivos CSV.
import numpy as np  # Computação numérica com arrays e operações matemáticas otimizadas.

# Bibliotecas para visualização de dados
import matplotlib.pyplot as plt  # Criação de gráficos e visualizações de dados.
import seaborn as sns  # Visualizações estatísticas avançadas.

# Biblioteca para pré-processamento de dados
from sklearn import preprocessing  # Funções para normalização, codificação e outras transformações de dados.
from sklearn.decomposition import PCA  # Análise de Componentes Principais para redução de dimensionalidade.
from sklearn.preprocessing import StandardScaler  # Normalização dos dados.

# Biblioteca para divisão e validação dos dados
import warnings  # Ignorar avisos que não são críticos para execução, como FutureWarnings
from sklearn.model_selection import train_test_split, cross_val_score  # Funções para dividir dados e realizar validação cruzada.

# Bibliotecas de modelos de Machine Learning (classificadores)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Bibliotecas de modelos de aprendizado não supervisionado
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

# Bibliotecas para cálculo de métricas de qualidade dos clusters
from sklearn.metrics import adjusted_rand_score  # Importa função para medir correspondência entre clusters e rótulos

# Bibliotecas para cálculo de métricas de desempenho dos modelos
from sklearn.metrics import silhouette_score
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error, r2_score, confusion_matrix



# Leitura do CSV
data = pd.read_csv('/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv')

print('-----------------------------------------------------------------------------------------------------')
print(f'PRIMEIROS 5 REGISTROS DO DATASET DEFINIDO:\n\n{data.head()}\n') # Visão rápida do dataset definido (5 primeiros registros)
print('-----------------------------------------------------------------------------------------------------')
```

```python
# DataShape (dimensões do DataFrame): número de linhas (entradas) por número de colunas (variáveis)
print(f'Data Shape (entradas, variáveis): {data.shape}')

# Averiguando infinitos / duplicados / nulos
print(f'Quantidade de valores infinitos: {(data == np.inf).sum().sum() + (data == -np.inf).sum().sum()}')
print(f'Quantidade de valores duplicados: {data.duplicated().sum()}')
print(f'Quantidade de registros com valor nulo: {data.isnull().sum().sum()}')
if (data.isnull().sum().sum() != 0):
    colunas_nulas = data.isnull().sum().sum()
    print('Colunas com valores nulos: \n' + colunas_nulas[colunas_nulas > 0])

# Retorna os dados: Número de linhas; número de colunas; nome das colunas; contagem de não-nulos de cada coluna; tipo de dado de cada coluna; uso de memória
print('\n-----------------------------------------------------------------------------------------------------')
print(f'DATA INFO:\n')
data.info()

# count -> Número de valores não-nulas de cada coluna
# mean -> Média aritmética dos valores de cada coluna (para colunas com valores numéricos)
# std -> Desvio padrão (medida de dispersão dos dados) de cada coluna, para colunas com valores numéricos
# min -> Mínimo (menor valor encontrado) para cada coluna, para colunas com valores numéricos
# 25% / 50% (mediana) / 75% -> Percentis (indicadores de distribuição dos dados)
# max -> Máximo (Maior valor encontrado) de cada coluna, para colunas com valores munéricos
print('\n-----------------------------------------------------------------------------------------------------')
print(f'DATA DESCRIBE:\n\n{data.describe()}')
print('\n-----------------------------------------------------------------------------------------------------')

# Ignora avisos de FutureWarning
with warnings.catch_warnings():
    warnings.simplefilter("ignore", FutureWarning)
    
    # Substitui valores infinitos por NaN
    data.replace([np.inf, -np.inf], np.nan, inplace=True)

    # Criação de um mapa de calor (heatmap)
    plt.figure(figsize=(16,10))
    sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
    plt.show()

    # Visualização da distribuição dos valores por variável
    for column in data.columns:
        plt.figure(figsize=(4, 1))
        sns.histplot(data[column], bins=20, kde=True, color='blue')
        plt.title(f'Distribuição dos Registros de {column}')
        plt.xlabel(column)
        plt.ylabel('Frequência')
        plt.show()
```

# Aplicação de machine learning

## Aprendizado supervisionado

```python
# Definindo x e y
x = data[['cp', 'thalachh', 'slp', 'oldpeak', 'exng']] #Variáveis cujos valores serão utilizadas para previsão
y = data['output'] #Variável cujos valores serão previstos

# Dividindo o conjunto de dados em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Aplicando diferentes modelos
modelos = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'Random Forest Classifier': RandomForestClassifier(n_estimators=50, max_depth=5, random_state=42),
    'Decision Tree Classifier': DecisionTreeClassifier(max_depth=5, random_state=42),
    'K-Nearest Neighbors (KNN)': KNeighborsClassifier(n_neighbors=5),
    'Support Vector Classifier (SVC)': SVC(random_state=42),
    'Rede Neural (MLP Classifier)': MLPClassifier(max_iter=1000, random_state=42)
}

for nome, modelo in modelos.items():
    print('\n-----------------------------------------------------------------------------------------------------')
    print(f'MODELO: {nome}') # Visão rápida do dataset definido (5 primeiros registros)
    print('-----------------------------------------------------------------------------------------------------\n')

    # Ajustando o modelo aos dados de treino
    modelo.fit(x_train, y_train)

    # Fazendo previsões com os dados de teste
    y_pred = modelo.predict(x_test)
    
    # Métricas
    
    accuracy = accuracy_score(y_test, y_pred)
    
    conf_matrix = confusion_matrix(y_test, y_pred)
    acertou_zero = conf_matrix[0, 0]  # Previu 0 corretamente
    acertou_um = conf_matrix[1, 1]  # Previu 1 corretamente
    errou_um = conf_matrix[1, 0]  # Previu 0 quando era 1
    errou_zero = conf_matrix[0, 1]  # Previu 1 quando era 0
    
    report = classification_report(y_test, y_pred, output_dict=True)
    precision_0 = report['0']['precision']
    recall_0 = report['0']['recall']
    f1_0 = report['0']['f1-score']
    precision_1 = report['1']['precision']
    recall_1 = report['1']['recall']
    f1_1 = report['1']['f1-score']
    
    # Imprimindo resultados das métricas
    
    print(f'Acurácia geral: {accuracy}')    
    print(f'\nMatriz de confusão:')    
    print(f'- Previu 0 corretamente (TN): {acertou_zero}')
    print(f'- Previu 1 corretamente (TP): {acertou_um}')
    print(f'- Previu 0 quando era 1 (FN): {errou_um}')
    print(f'- Previu 1 quando era 0 (FP): {errou_zero}')
    
    print(f'\nRelatório de Classificação:')
    print(f'- Precisão para classe 0: {precision_0:.2f}')
    print(f'- Recall para classe 0: {recall_0:.2f}')
    print(f'- F1-score para classe 0: {f1_0:.2f}')
    print(f'- Precisão para classe 1: {precision_1:.2f}')
    print(f'- Recall para classe 1: {recall_1:.2f}')
    print(f'- F1-score para classe 1: {f1_1:.2f}')
```

```python
# Fazendo previsões com valores inputados
# Modelos com melhor desempenho: Logistic Regression e Rede Neural

# Dados de entrada para a previsão
input_data = {'cp': 3, 'thalachh': 187, 'slp': 2, 'oldpeak': 2.3, 'exng': 1}  # 'cp', 'thalachh', 'slp', 'oldpeak', 'exng'
input_data = pd.DataFrame([input_data])

print('\n-----------------------------------------------------------------------------------------------------')
print('PREVISÃO COM VALORES INPUTADOS PARA 2 MODELOS COM MELHOR DESEMPENHO')
print('-----------------------------------------------------------------------------------------------------\n')

# Fazendo a previsão com Logistic Regression (usando o modelo já treinado)
predicted_output_lr = modelos['Logistic Regression'].predict(input_data)[0]
print(f"Previsão com Logistic Regression: {predicted_output_lr} ({'maior' if predicted_output_lr == 1 else 'menor'} chance de ataque cardíaco)")

# Fazendo a previsão com Rede Neural (usando o modelo já treinado)
predicted_output_mlp = modelos['Rede Neural (MLP Classifier)'].predict(input_data)[0]
print(f"Previsão com Rede Neural (MLP Classifier): {predicted_output_mlp} ({'maior' if predicted_output_mlp == 1 else 'menor'} chance de ataque cardíaco)\n")

# Comparando os resultados das previsões dos dois modelos
if predicted_output_lr == predicted_output_mlp:
    # Caso os modelos realizem mesma previsão
    print(f"Os dois modelos realizaram a mesma previsão: {predicted_output_lr} ({'maior' if predicted_output_lr == 1 else 'menor'} chance)")
else:
    # Caso os modelos realizem previsões distintas
    print(f"Os dois modelos realizaram previsões distintas: Logistic Regression previu {predicted_output_lr} ({'maior' if predicted_output_lr == 1 else 'menor'} chance), enquanto Rede Neural previu {predicted_output_mlp} ({'maior' if predicted_output_mlp == 1 else 'menor'} chance)")

print('\n-----------------------------------------------------------------------------------------------------')
print('PREVISÃO COM VALORES INPUTADOS PARA TODOS OS MODELOS')
print('-----------------------------------------------------------------------------------------------------')

# Fazendo previsões com todos os modelos usando os valores inputados
resultados_previsao = {}
for nome, modelo in modelos.items():
    predicted_output = modelo.predict(input_data)[0]
    resultados_previsao[nome] = predicted_output

# Comparando os resultados das previsões entre todos os modelos
previsoes_0 = len([previsao for previsao in resultados_previsao.values() if previsao == 0]) # Contador de modelos que previram 0
previsoes_1 = len([previsao for previsao in resultados_previsao.values() if previsao == 1]) # Contador de modelos que previram 1
total_modelos = len(modelos)  # Contador do total de modelos

# Exibindo o resumo das previsões
print(f"\nResumo das previsões: {previsoes_0}/{total_modelos} modelos previram 0 (menor chance), {previsoes_1}/{total_modelos} modelos previram 1 (maior chance)")

# Verificando se todos os modelos realizaram a mesma previsão
if previsoes_0 == total_modelos or previsoes_1 == total_modelos:
    # Caso todos os modelos realizem mesma previsão
    print(f"Todos os modelos realizaram a mesma previsão: {1 if previsoes_1 == total_modelos else 0} ({'maior' if previsoes_1 == total_modelos else 'menor'} chance de ataque cardíaco)")
else:
    # Caso haja previsões diferentes para os modelos
    print("\nComparação das previsões entre todos os modelos:")
    for nome, previsao in resultados_previsao.items():
        print(f"{nome} previu {previsao} ({'maior' if previsao == 1 else 'menor'} chance de ataque cardíaco)")
```

## Aprendizado não supervisionado

```python
# Essa é uma experimentação inicial com modelos de aprendizado não supervisionado. Os modelos não obtiveram o melhor desempenho, provavelmente necessitando de maiores ajustes nos dados ou da aplicação de modelos mais adequados
print(
    '''
    <<< IMPORTANTE! Essa é uma experimentação inicial com modelos de aprendizado não supervisionado. Os modelos não obtiveram o melhor desempenho, provavelmente necessitando de maiores ajustes nos dados ou da aplicação de modelos mais adequados >>>
    '''
)

# Pré-processamento: Normalização dos dados
scaler = StandardScaler()  # Inicializa o escalador para normalizar os dados
x_scaled = scaler.fit_transform(x)  # Normaliza os dados (z-score)

print('\n-----------------------------------------------------------------------------------------------------')
print('K-MEANS CLUSTERING')
print('-----------------------------------------------------------------------------------------------------\n')

pca = PCA(n_components=2)  # Redução de dimensionalidade para visualização
x_pca = pca.fit_transform(x_scaled)  # Transforma os dados em duas dimensões para facilitar a visualização

kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)  # Inicializando K-Means com n_init definido explicitamente para evitar warnings
clusters_kmeans = kmeans.fit_predict(x_pca)  # Ajusta o modelo K-Means aos dados e prediz os clusters
print(f'Clusters atribuídos pelo K-Means:\n{clusters_kmeans}\n')

# Interpretação dos clusters atribuídos pelo K-Means
unique, counts = np.unique(clusters_kmeans, return_counts=True)  # Conta quantos pontos pertencem a cada cluster
kmeans_cluster_distribution = dict(zip(unique, counts))  # Cria um dicionário para armazenar a distribuição dos clusters
print('Distribuição dos pontos entre os clusters K-Means:')
for cluster, count in kmeans_cluster_distribution.items():
    print(f'- Cluster {cluster}: {count} pontos')

# Validando a qualidade dos clusters formados pelo K-Means
kmeans_ari = adjusted_rand_score(y, clusters_kmeans)  # Calcula o ARI (Ajusted Rand Index) para medir a correspondência com os rótulos verdadeiros
print(f'\nAjusted Rand Index (ARI) para K-Means: {kmeans_ari:.2f} (quanto mais próximo de 1, melhor a correspondência com os rótulos reais)')

# Calculando o Silhouette Score
kmeans_silhouette = silhouette_score(x_pca, clusters_kmeans)  # Calcula o Silhouette Score para avaliar a coesão dos clusters
print(f'Silhouette Score para K-Means: {kmeans_silhouette:.2f} (quanto mais próximo de 1, melhor a coesão dos clusters)')

print('\n-----------------------------------------------------------------------------------------------------')
print('GAUSSIAN MIXTURE MODEL (GMM)')
print('-----------------------------------------------------------------------------------------------------\n')

gmm = GaussianMixture(n_components=2, random_state=42)  # Inicializa o modelo GMM com 2 componentes (clusters)
gmm.fit(x_pca)  # Ajusta o modelo GMM aos dados de entrada
clusters_gmm = gmm.predict(x_pca)  # Prediz os clusters a que cada ponto pertence
print(f'Clusters atribuídos pelo GMM:\n{clusters_gmm}\n')

# Interpretação dos clusters atribuídos pelo GMM
unique, counts = np.unique(clusters_gmm, return_counts=True)  # Conta quantos pontos pertencem a cada cluster
gmm_cluster_distribution = dict(zip(unique, counts))  # Cria um dicionário para armazenar a distribuição dos clusters
print('Distribuição dos pontos entre os clusters GMM:')
for cluster, count in gmm_cluster_distribution.items():
    print(f'- Cluster {cluster}: {count} pontos')

# Validando a qualidade dos clusters formados pelo GMM
gmm_ari = adjusted_rand_score(y, clusters_gmm)  # Calcula o ARI (Ajusted Rand Index) para medir a correspondência com os rótulos verdadeiros
print(f'\nAjusted Rand Index (ARI) para GMM: {gmm_ari:.2f} (quanto mais próximo de 1, melhor a correspondência com os rótulos reais)')

# Calculando o Silhouette Score
gmm_silhouette = silhouette_score(x_pca, clusters_gmm)  # Calcula o Silhouette Score para avaliar a coesão dos clusters
print(f'Silhouette Score para GMM: {gmm_silhouette:.2f} (quanto mais próximo de 1, melhor a coesão dos clusters)')
```

```python
print('\n-----------------------------------------------------------------------------------------------------')
print('VERIFICAÇÃO DOS CLUSTERS ATRIBUÍDOS PELOS MODELOS NÃO SUPERVISIONADOS (K-MEANS)')
print('-----------------------------------------------------------------------------------------------------\n')

# Verificando correspondência dos clusters do K-Means com as classes reais
y_cluster_0 = y[clusters_kmeans == 0]
y_cluster_1 = y[clusters_kmeans == 1]

# Contando quantos valores 0 e 1 há em cada cluster do K-Means
count_0_cluster_0 = (y_cluster_0 == 0).sum()
count_1_cluster_0 = (y_cluster_0 == 1).sum()
count_0_cluster_1 = (y_cluster_1 == 0).sum()
count_1_cluster_1 = (y_cluster_1 == 1).sum()

print(f"Cluster 0 (K-Means): {count_0_cluster_0} casos de baixa chance, {count_1_cluster_0} casos de alta chance")
print(f"Cluster 1 (K-Means): {count_0_cluster_1} casos de baixa chance, {count_1_cluster_1} casos de alta chance")

# Determinando qual cluster do K-Means representa "baixa" e "alta" chance com base na maioria
cluster_baixa_chance_kmeans = 0 if count_0_cluster_0 > count_1_cluster_0 else 1
cluster_alta_chance_kmeans = 1 if cluster_baixa_chance_kmeans == 0 else 0

print(f"\nCluster {cluster_baixa_chance_kmeans} (K-Means) é predominantemente 'baixa chance' de ataque cardíaco")
print(f"Cluster {cluster_alta_chance_kmeans} (K-Means) é predominantemente 'alta chance' de ataque cardíaco")

print('\n-----------------------------------------------------------------------------------------------------')
print('VERIFICAÇÃO DOS CLUSTERS ATRIBUÍDOS PELOS MODELOS NÃO SUPERVISIONADOS (GMM)')
print('-----------------------------------------------------------------------------------------------------\n')

# Verificando correspondência dos clusters do GMM com as classes reais
y_cluster_0_gmm = y[clusters_gmm == 0]
y_cluster_1_gmm = y[clusters_gmm == 1]

# Contando quantos valores 0 e 1 há em cada cluster do GMM
count_0_cluster_0_gmm = (y_cluster_0_gmm == 0).sum()
count_1_cluster_0_gmm = (y_cluster_0_gmm == 1).sum()
count_0_cluster_1_gmm = (y_cluster_1_gmm == 0).sum()
count_1_cluster_1_gmm = (y_cluster_1_gmm == 1).sum()

print(f"Cluster 0 (GMM): {count_0_cluster_0_gmm} casos de baixa chance, {count_1_cluster_0_gmm} casos de alta chance")
print(f"Cluster 1 (GMM): {count_0_cluster_1_gmm} casos de baixa chance, {count_1_cluster_1_gmm} casos de alta chance")

# Determinando qual cluster do GMM representa "baixa" e "alta" chance com base na maioria
cluster_baixa_chance_gmm = 0 if count_0_cluster_0_gmm > count_1_cluster_0_gmm else 1
cluster_alta_chance_gmm = 1 if cluster_baixa_chance_gmm == 0 else 0

print(f"\nCluster {cluster_baixa_chance_gmm} (GMM) é predominantemente 'baixa chance' de ataque cardíaco")
print(f"Cluster {cluster_alta_chance_gmm} (GMM) é predominantemente 'alta chance' de ataque cardíaco")
```

```python
# Previsão com valores inputados
print('\n-----------------------------------------------------------------------------------------------------')
print('PREVISÃO COM VALORES INPUTADOS PARA MODELOS NÃO SUPERVISIONADOS')
print('-----------------------------------------------------------------------------------------------------\n')

# Dados de entrada para a previsão
input_data_ns = {'cp': 3, 'thalachh': 187, 'slp': 2, 'oldpeak': 2.3, 'exng': 1}  # 'cp', 'thalachh', 'slp', 'oldpeak', 'exng'
input_data_ns = pd.DataFrame([input_data_ns])

# Aplicando a normalização aos dados de entrada
input_data_scaled = scaler.transform(input_data_ns)

# Aplicando o PCA aos dados de entrada
input_data_pca = pca.transform(input_data_scaled)

# Previsão com K-Means
input_cluster_kmeans = kmeans.predict(input_data_pca)[0]
resultado_kmeans = 'maior chance de ataque cardíaco' if input_cluster_kmeans == cluster_alta_chance_kmeans else 'menor chance de ataque cardíaco'
print(f"Previsão com K-Means: Cluster {input_cluster_kmeans} ({resultado_kmeans})")

# Previsão com GMM
input_cluster_gmm = gmm.predict(input_data_pca)[0]
resultado_gmm = 'maior chance de ataque cardíaco' if input_cluster_gmm == cluster_alta_chance_gmm else 'menor chance de ataque cardíaco'
print(f"Previsão com GMM: Cluster {input_cluster_gmm} ({resultado_gmm})")

# Comparando os resultados das previsões dos dois modelos
if input_cluster_kmeans == input_cluster_gmm:
    # Caso os modelos realizem a mesma previsão
    print(f"\nOs dois modelos não supervisionados realizaram a mesma previsão: {resultado_kmeans}")
else:
    # Caso os modelos realizem previsões distintas
    print(f"\nOs dois modelos não supervisionados realizaram previsões distintas: "
          f"K-Means previu {resultado_kmeans}, enquanto GMM previu {resultado_gmm}")
```
